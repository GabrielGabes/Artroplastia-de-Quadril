# PRÉ PROCESSAMENTO

```python
# ASSA
df['assa'] = df['asa']
df['assa'] = df['assa'].replace({4:3}, regex=True)

#GERALPERI
df.drop(df[df['tecnica'] == 'geralperi'].index, inplace=True)

#GERALL
lista_gerall = []
for i in df['geral']:
    if i == 'venosa':
        lista_gerall.append(1)
    elif i == 'inalatoria':
        lista_gerall.append(2)
    else:
        lista_gerall.append(3)
df['gerall'] = lista_gerall

#LIQUIDO
df['liq'] = (df['líquidos total']/df['peso'])/(df['tempo cirúrgico (min)']/60)

#IDADE
lista_idade = []
for i in df['idade']:
    if i < 50:
        lista_idade.append(0)
    if i >= 50 and i < 60:
        lista_idade.append(1)
    if i >= 60 and i < 70:
        lista_idade.append(2)
    if i >= 70 and i < 80:
        lista_idade.append(3)
    if i >= 80 and i <120: #else: lista_idade.append(3)
        lista_idade.append(3)
df['id'] = lista_idade

#PESO - OBESO - SOBREPESO
lista_obeso = []
for i in df['peso']:
    if i < 100:
        lista_obeso.append(1)
    if i >= 100 and i <140: #else: lista_obeso.append(2)
        lista_obeso.append(2)
df['obeso'] = lista_obeso


#PRIORDAY - 
df['TEMPO (DIAS) ANTES DA CIRURGIA'].fillna(1, inplace = True) #removendo valor n.a
lista_priorday = []
for i in df['TEMPO (DIAS) ANTES DA CIRURGIA']:
    if i <= 1:
        lista_priorday.append(1)
    if i > 1 and int(i) <= 30:
        lista_priorday.append(2)
df['priorday'] = lista_priorday

#HIPOTENSÃO
lista_hipo = []
for i in df['hipotensão no IO > de 20%?']:
    if i == 'não':
        lista_hipo.append(0)
    if i == 'sim':
        lista_hipo.append(1)
df['hipo'] = lista_hipo
```

# Função Tabela de Contingência

```python
def tabela_cont(col1, col2):
    a = df.groupby([col1, col2]).size().unstack()
    for i in a.columns:
        a[i].fillna(0, inplace = True)
        a[i] = a[i].apply(lambda x: int(x))
    display(a)
    print('='*100)
```

```python
#Biblioteca que vai possibilitar transformação de variavel categorico para numerico
from sklearn.preprocessing import LabelEncoder

#Instanciando um objeto para cada atributo categorico
label_encoder_Ano = LabelEncoder()
label_encoder_sexo = LabelEncoder()
label_encoder_AVC = LabelEncoder()
label_encoder_tabagismo = LabelEncoder()
label_encoder_doenca_cardiaca = LabelEncoder()
label_encoder_doenca_dm = LabelEncoder()
label_encoder_doenca_has = LabelEncoder()
label_encoder_doenca_DRC = LabelEncoder()
label_encoder_doenca_tecnica = LabelEncoder()
label_encoder_doenca_tipo = LabelEncoder()
label_encoder_doenca_DVA = LabelEncoder()
label_encoder_doenca_sangramento = LabelEncoder()
label_encoder_doenca_Transamin = LabelEncoder()
label_encoder_doenca_transfusao = LabelEncoder()
label_encoder_doenca_hipotermia = LabelEncoder()
label_encoder_doenca_prot_hipo = LabelEncoder()

#Aplicando o processamento
X_df[:,0] = label_encoder_Ano.fit_transform(X_df[:,0])
X_df[:,2] = label_encoder_sexo.fit_transform(X_df[:,2])
X_df[:,5] = label_encoder_AVC.fit_transform(X_df[:,5])
X_df[:,6] = label_encoder_tabagismo.fit_transform(X_df[:,6])
X_df[:,7] = label_encoder_doenca_cardiaca.fit_transform(X_df[:,7])
X_df[:,10] = label_encoder_doenca_dm.fit_transform(X_df[:,10])
X_df[:,11] = label_encoder_doenca_has.fit_transform(X_df[:,11])
X_df[:,12] = label_encoder_doenca_DRC.fit_transform(X_df[:,12])
X_df[:,13] = label_encoder_doenca_tecnica.fit_transform(X_df[:,13])
X_df[:,15] = label_encoder_doenca_tipo.fit_transform(X_df[:,15])
X_df[:,16] = label_encoder_doenca_DVA.fit_transform(X_df[:,16])
X_df[:,17] = label_encoder_doenca_sangramento.fit_transform(X_df[:,17])
X_df[:,18] = label_encoder_doenca_Transamin.fit_transform(X_df[:,18])
X_df[:,19] = label_encoder_doenca_transfusao.fit_transform(X_df[:,19])
X_df[:,20] = label_encoder_doenca_hipotermia.fit_transform(X_df[:,20])
X_df[:,21] = label_encoder_doenca_prot_hipo.fit_transform(X_df[:,21])


X_df #transformação feita
```

# Implementando Algoritmo de Machine Learning

```python
from sklearn.model_selection import train_test_split
X_treino, X_teste, y_treino, y_teste = train_test_split(X_df, y_df, test_size=0.25, random_state=0)

X_treino.shape, y_treino.shape, X_teste.shape, y_teste.shape
```

# ARVORE DE DECISÃO

```python
from sklearn.tree import DecisionTreeClassifier

classificador_arvore = DecisionTreeClassifier(criterion='entropy', random_state = 0)
classificador_arvore.fit(X_treino, y_treino)

previsoes = classificador_arvore.predict(X_teste)

#Comparativo
display(previsoes)
display(y_teste)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

accuracy_arvore = accuracy_score(y_teste, previsoes)
print('Acuraria-->', accuracy_arvore)
```

```python
#Simples:
matriz_confusao = confusion_matrix(y_teste, previsoes)
display(matriz_confusao)

from yellowbrick.classifier import ConfusionMatrix
cm = ConfusionMatrix(classificador_arvore)
cm.fit(X_treino, y_treino)
cm.score(X_teste, y_teste)
```

```python
from sklearn.metrics import classification_report
Previsao = classification_report(y_teste, previsoes)
print(Previsao)
```

```python
#visualizar a arvore de decisão
from sklearn import tree
previsores = ['Ano','id','sexo','assa','ELETIVA /URGENCIA','AVC prévio','tabagismo','Doença cardíaca','obeso','priorday','dm','has','DRC','tecnica','gerall','tipo','DVA','sangramento','Transamin','transfusão','hipotermia','protocolo de hipotermia','liq']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (100,100)) #sup-grafico
tree.plot_tree(classificador_arvore, feature_names=previsores, class_names=['0','1'], filled=True);
#fig.savefig('arvore_credit.png')
```
