```python
import pandas as pd #data frames
import numpy as np #matematica

import matplotlib.pyplot as plt #base grafico
%matplotlib inline

import seaborn as sns #graficos
import datetime as dt #datas

df = pd.read_excel('(BASE INALTERADA)BANCO QUADRIL - dezembro 2021.xlsx')
df.head().T
```

# 1 - EXPLORAÇÃO e PRÉ PROCESSAMENTOS

```python
(df.columns)
(df.shape)
```

```python
import unidecode

news_columns = []
for i in df.columns:
    news_columns.append(unidecode.unidecode(i.strip().replace(' ', '_').lower()))

df.columns = news_columns
df.columns
```

```python
#excluindo colunas inuteis 
df = df.drop(columns=['atendimento', 'comentarios', 'hospital'])

#df['Doença cardíaca'].unique()
#valores repedidos
df['doenca_cardiaca'] = df['doenca_cardiaca'].replace({'Sim':'sim'},regex=True)

#ASSA
df['asa'] = df['asa'].replace({4:3}, regex=True)

#GERALPERI
df.drop(df[df['tecnica'] == 'geralperi'].index, inplace=True)

#LIQUIDO
df['liq'] = (df['liquidos_total']/df['peso'])/(df['tempo_cirurgico_(min)']/60)

df['priorday'] = df['tempo_(dias)_antes_da_cirurgia']

#MUDANDO O NOME DA COLUNA
df.rename(columns={'hipotensao_no_io_>_de_20%?':'hipo',
                  'eletiva_/urgencia':'eletiva_urgencia',
                  'tempo_cirurgico_(min)':'tempo_cirurgico_min'}, inplace = True)
```

```python
round(df.describe().T, 2)
```

```python
#PRIORDAY
lista = []
for i in df['priorday']:
    if i <= 1:
        lista.append(0)
    elif i > 1:
        lista.append(1)
    else:
        lista.append(np.nan)
df['priorday'] = lista

df['priorday'].value_counts()
```

```python
df['eletiva_urgencia'].unique()
```

```python
df = df.reset_index(drop=True)
lista = []

for i in range(len(df)):
    var1 = df['eletiva_urgencia'][i]
    var2 = df['priorday'][i]
    
    if var1 == 1:
        lista.append(0)
    elif (var1 == 2) and (var2 == 1):
        lista.append(1)
    else: 
        lista.append(2)
                
print(len(lista))

df['tempo'] = lista

a = df.groupby(['tempo'])['tempo'].count().reset_index(name='counts')
display(a)
```

```python
#IDADE
lista = []
for i in df['idade']:
    if i < 50:
        lista.append(0) #'<50'
    elif i >= 50 and i < 60:
        lista.append(1) #'50-59'
    elif i >= 60 and i < 70:
        lista.append(2) #'60-69'
    elif i >= 70 and i < 80:
        lista.append(3) #'70-79'
    elif i >= 80:
        lista.append(4) #'<80'
    else:
        lista.append(np.nan)
df['id'] = lista
'''a = df.groupby(['id'])['id'].count().reset_index(name='fi')
a['fr'] = (a['fi']*100)/sum(a['fi'])
display(a)'''
```

```python
#PESO
lista = []
for i in df['peso']:
    if i < 100: #... and i < 120
        lista.append(0)
    elif i >= 100:
        lista.append(1)
    else: 
        lista.append(np.nan)

df['obeso'] = lista
```

```python
#TIM
lista = []
for i in df['tempo_de_internacao']:
    if i > 5: #... and i < 120
        lista.append(1)
    elif i <= 5:
        lista.append(0)
    else: 
        lista.append(np.nan)

df['tim'] = lista
```

```python
#--------TESTE 1
colunas = ['idade', 'peso', 'liq']

for col in colunas:
    lista = []
    for i in df[col]:
        if i <= np.percentile(df[col], 25):
            lista.append('<=Q1')
        elif i > np.percentile(df[col], 25) and i <= np.percentile(df[col], 50):
            lista.append('<Q2<=')
        elif i > np.percentile(df[col], 50) and i <= np.percentile(df[col], 75):
            lista.append('<Q3<=')
        else:
            lista.append('<Q4')
    df[str(col)+'_q'] = lista


for i in ['idade_q', 'peso_q', 'liq_q']:
    a = df.groupby([i, 'tecnica']).size().unstack()
    for t in a.columns:
        a[t].fillna(0, inplace=True)
        a[t] = a[t].apply(lambda x: int(x))
    a.loc['total'] = a.sum(axis=0)
    for c in a.columns:
        a[c] = a[c]*100/a[c]['total']
    display(round(a, 2))
    
display(df[['idade', 'peso', 'liq']].describe().T)
```

```python
X_df = hipo.iloc[:, 0:23].values
y_df = hipo.iloc[:, 23].values

#display(X_df)
#display(y_df)
```

```python
#Biblioteca que vai possibilitar transformação de variavel categorico para numerico
from sklearn.preprocessing import LabelEncoder

#Instanciando um objeto para cada atributo categorico
#Aplicando o processamento

label_encoder_Ano = LabelEncoder()
X_df[:,0] = label_encoder_Ano.fit_transform(X_df[:,0])
label_encoder_sexo = LabelEncoder()
X_df[:,2] = label_encoder_sexo.fit_transform(X_df[:,2])
label_encoder_AVC_previo = LabelEncoder()
X_df[:,5] = label_encoder_AVC_previo.fit_transform(X_df[:,5])
label_encoder_tabagismo = LabelEncoder()
X_df[:,6] = label_encoder_tabagismo.fit_transform(X_df[:,6])
label_encoder_Doenca_cardiaca = LabelEncoder()
X_df[:,7] = label_encoder_Doenca_cardiaca.fit_transform(X_df[:,7])
label_encoder_dm = LabelEncoder()
X_df[:,10] = label_encoder_dm.fit_transform(X_df[:,10])
label_encoder_has = LabelEncoder()
X_df[:,11] = label_encoder_has.fit_transform(X_df[:,11])
label_encoder_DRC = LabelEncoder()
X_df[:,12] = label_encoder_DRC.fit_transform(X_df[:,12])
label_encoder_tecnica = LabelEncoder()
X_df[:,13] = label_encoder_tecnica.fit_transform(X_df[:,13])
label_encoder_geral = LabelEncoder()
X_df[:,14] = label_encoder_geral.fit_transform(X_df[:,14])
label_encoder_tipo = LabelEncoder()
X_df[:,15] = label_encoder_tipo.fit_transform(X_df[:,15])
label_encoder_DVA = LabelEncoder()
X_df[:,16] = label_encoder_DVA.fit_transform(X_df[:,16])
label_encoder_sangramento = LabelEncoder()
X_df[:,17] = label_encoder_sangramento.fit_transform(X_df[:,17])
label_encoder_Transamin = LabelEncoder()
X_df[:,18] = label_encoder_Transamin.fit_transform(X_df[:,18])
label_encoder_transfusão = LabelEncoder()
X_df[:,19] = label_encoder_transfusão.fit_transform(X_df[:,19])
label_encoder_hipotermia = LabelEncoder()
X_df[:,20] = label_encoder_hipotermia.fit_transform(X_df[:,20])
label_encoder_protocolo_de_hipotermia = LabelEncoder()
X_df[:,21] = label_encoder_protocolo_de_hipotermia.fit_transform(X_df[:,21])

X_df
```

```python
from sklearn.model_selection import train_test_split
X_treino, X_teste, y_treino, y_teste = train_test_split(X_df, y_df, test_size=0.25, random_state=0)

X_treino.shape, y_treino.shape, X_teste.shape, y_teste.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

classificador_arvore = DecisionTreeClassifier(criterion='entropy', random_state = 0)
classificador_arvore.fit(X_treino, y_treino)

previsoes = classificador_arvore.predict(X_teste)

#Comparativo
#display(previsoes)
#display(y_teste)

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

accuracy_arvore = accuracy_score(y_teste, previsoes)
print('Acuraria-->', accuracy_arvore)
```
