```python
import pandas as pd #data frames
import numpy as np #matematica

import matplotlib.pyplot as plt #base grafico
%matplotlib inline
```

```python
df = pd.read_csv('df.csv')
df.head().T
```

# Tabela 1. Características dos pacientes submetidos a anestesia para artroplastia de quadril de janeiro de 2016 a dezembro de 2021.

```python
for i in df[['id','ano','sexo','asa','tabagismo','avc_previo','doenca_cardiaca','has','dm',
            'obeso','tecnica','tipo','sangramento','dva','transfusao','transamin','hipotermia',
             'drc','hipo','destino']]:
    a = df.groupby([i])[i].count().reset_index(name='fi')
    a['fr'] = round((a['fi']*100)/sum(a['fi']),2)
    display(a)
```

# Tabela 4. Características dos pacientes, evolução pós-operatória e técnica anestésica. 

```python
for i in df[['id','ano','sexo','asa','eletiva_urgencia','tabagismo','avc_previo',
            'doenca_cardiaca','has','dm','tipo','protocolo_de_hipotermia','sangramento',
            'dva','transfusao','transamin','hipotermia','drc','hipo','destino','necessidade_de_dva',
            'dor_24h','pcr','transfusao_na_uti','obeso','tipo','tempo','priorday','fii','hipo']]:
    a = df.groupby([i, 'tecnica']).size().unstack()
    for t in a.columns:
        a[t].fillna(0, inplace=True)
        a[t] = a[t].apply(lambda x: int(x))
    #a.loc['total'] = a.sum(axis=0)
    #for c in a.columns:
        #a[c] = a[c]*100/a[c]['total']
    display(round(a, 2))
```

# tabela 5: Fazer box plot

```python
colunas = ['idade', 'peso', 'liq']

for col in colunas:
    lista = []
    for i in df[col]:
        if i <= np.percentile(df[col], 25):
            lista.append('<=Q1')
        elif i > np.percentile(df[col], 25) and i <= np.percentile(df[col], 50):
            lista.append('<Q2<=')
        elif i > np.percentile(df[col], 50) and i <= np.percentile(df[col], 75):
            lista.append('<Q3<=')
        else:
            lista.append('<Q4')
    df[str(col)+'_q'] = lista


for i in ['liq_q']:
    a = df.groupby([i, 'tecnica']).size().unstack()
    for t in a.columns:
        a[t].fillna(0, inplace=True)
        a[t] = a[t].apply(lambda x: int(x))
        b = a.copy()
    a.loc['total'] = a.sum(axis=0)
    for c in a.columns:
        a[c] = a[c]*100/a[c]['total']
    display(round(a, 2))
    display(round(b, 2))
```

```python
import seaborn as sns

plt.figure(figsize=(25,8))
plt.subplot(3,1,1)
sns.histplot(data=df['liq'])

plt.subplot(3,1,2)
sns.boxplot(x=df['liq'])

plt.subplot(3,1,3)
sns.kdeplot(data=df['liq']);
```
