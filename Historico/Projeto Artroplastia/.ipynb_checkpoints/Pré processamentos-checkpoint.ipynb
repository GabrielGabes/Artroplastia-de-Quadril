```python
import pandas as pd #data frames
import numpy as np #matematica

import matplotlib.pyplot as plt #base grafico
%matplotlib inline
```

```python
df = pd.read_excel('(BASE INALTERADA)BANCO QUADRIL - dezembro 2021.xlsx')
df.head().T
```

```python
import unidecode

news_columns = []
for i in df.columns:
    news_columns.append(unidecode.unidecode(i.strip().replace(' ', '_').lower()))

df.columns = news_columns
df.columns
```

```python
#excluindo colunas inuteis 
df = df.drop(columns=['atendimento', 'comentarios', 'hospital'])

#df['Doença cardíaca'].unique()
#valores repedidos
df['doenca_cardiaca'] = df['doenca_cardiaca'].replace({'Sim':'sim'},regex=True)

#ASSA
df['asa'] = df['asa'].replace({4:3}, regex=True)

#GERALPERI
df.drop(df[df['tecnica'] == 'geralperi'].index, inplace=True)

#LIQUIDO
df['liq'] = (df['liquidos_total']/df['peso'])/(df['tempo_cirurgico_(min)']/60)

df['priorday'] = df['tempo_(dias)_antes_da_cirurgia']

#MUDANDO O NOME DA COLUNA
df.rename(columns={'hipotensao_no_io_>_de_20%?':'hipo',
                  'eletiva_/urgencia':'eletiva_urgencia',
                  'tempo_cirurgico_(min)':'tempo_cirurgico_min'}, inplace = True)
```

```python
#PRIORDAY
lista = []
for i in df['priorday']:
    if i <= 1:
        lista.append(0)
    elif i > 1:
        lista.append(1)
    else:
        lista.append(np.nan)
df['priorday'] = lista

df['priorday'].value_counts()
```

```python
#TEMPO
df = df.reset_index(drop=True)
lista = []

for i in range(len(df)):
    var1 = df['eletiva_urgencia'][i]
    var2 = df['priorday'][i]
    
    if var1 == 1:
        lista.append(0)
    elif (var1 == 2) and (var2 == 1):
        lista.append(1)
    else: 
        lista.append(2)
                
print(len(lista))

df['tempo'] = lista

a = df.groupby(['tempo'])['tempo'].count().reset_index(name='counts')
display(a)
```

```python
#IDADE
lista = []
for i in df['idade']:
    if i < 50:
        lista.append(0) #'<50'
    elif i >= 50 and i < 60:
        lista.append(1) #'50-59'
    elif i >= 60 and i < 70:
        lista.append(2) #'60-69'
    elif i >= 70 and i < 80:
        lista.append(3) #'70-79'
    elif i >= 80:
        lista.append(4) #'<80'
    else:
        lista.append(np.nan)
df['id'] = lista
'''a = df.groupby(['id'])['id'].count().reset_index(name='fi')
a['fr'] = (a['fi']*100)/sum(a['fi'])
display(a)'''
```

```python
#PESO
lista = []
for i in df['peso']:
    if i < 100: #... and i < 120
        lista.append(0)
    elif i >= 100:
        lista.append(1)
    else: 
        lista.append(np.nan)

df['obeso'] = lista
```

```python
#TIM
lista = []
for i in df['tempo_de_internacao']:
    if i > 5: #... and i < 120
        lista.append(1)
    elif i <= 5:
        lista.append(0)
    else: 
        lista.append(np.nan)

df['tim'] = lista
```

```python
df = df.reset_index(drop=True)

#'TAXA DE ADESAO AO PROTOCOLO'
lista = []

for i in range(len(df)):
    var1 = df['necessidade_de_dva'][i]
    var2 = df['transfusao_na_uti'][i]
    
    if (var1 == "não") and (var2 == "não"):
        lista.append(0)
    elif (var1 == "sim") or (var2 == "sim"):
        lista.append(1)
    else: 
        lista.append(np.nan)
                
print(len(lista))

df['fii'] = lista

a = df.groupby(['fii'])['fii'].count().reset_index(name='counts')
display(a)
```
